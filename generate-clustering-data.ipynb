```python
import pandas as pd
import seaborn as sns
```

# 2.0. Loading data

```python
df_exportacao = pd.read_csv('C:\\Users\\mathe\\Downloads\\Dados - GAMMA Challenge 2022\\Exportacoes (COMEXSTAT)\\exportacao_full.csv')
df_uf_mun = pd.read_excel('C:\\Users\\mathe\\Downloads\\Dados - GAMMA Challenge 2022\\lista_de_municipios_Amazonia_Legal_2020.xls')
```

```python
base_munic_2020_rh = pd.read_excel('C:\\Users\\mathe\\Downloads\\Dados - GAMMA Challenge 2022\\MUNIC\\Base_MUNIC_2020.xlsx', sheet_name=1)
base_munic_2020_habit = pd.read_excel('C:\\Users\\mathe\\Downloads\\Dados - GAMMA Challenge 2022\\MUNIC\\Base_MUNIC_2020.xlsx', sheet_name=2)
base_munic_2020_trans = pd.read_excel('C:\\Users\\mathe\\Downloads\\Dados - GAMMA Challenge 2022\\MUNIC\\Base_MUNIC_2020.xlsx', sheet_name=3)
base_munic_2020_agro = pd.read_excel('C:\\Users\\mathe\\Downloads\\Dados - GAMMA Challenge 2022\\MUNIC\\Base_MUNIC_2020.xlsx', sheet_name=4)
base_munic_2020_meio_amb = pd.read_excel('C:\\Users\\mathe\\Downloads\\Dados - GAMMA Challenge 2022\\MUNIC\\Base_MUNIC_2020.xlsx', sheet_name=5)
base_munic_2020_gest_riscos = pd.read_excel('C:\\Users\\mathe\\Downloads\\Dados - GAMMA Challenge 2022\\MUNIC\\Base_MUNIC_2020.xlsx', sheet_name=6)
```

## 2.1. Transforming data

```python
df_uf_mun['City'] = df_uf_mun['NM_MUN'].astype(str) + ' - ' +  df_uf_mun['SIGLA'].astype(str)
df_uf_mun.head()
```

## 2.2. Merging data

```python
df_raw = df_uf_mun[['CD_MUN', 'NM_MUN']].merge(base_munic_2020_rh.drop(['UF', 'Cod UF','Mun'], axis=1), how='inner', left_on='CD_MUN', right_on='CodMun')
df_raw = df_raw.merge(base_munic_2020_habit.drop(['UF', 'Cod UF', 'Mun', 'Faixa_pop', 'Regiao'], axis=1), how='inner', left_on='CD_MUN', right_on='CodMun')
df_raw = df_raw.merge(base_munic_2020_agro.drop(['UF', 'Cod UF','Mun', 'Faixa_pop', 'Regiao'], axis=1), how='inner', left_on='CD_MUN', right_on='CodMun')
df_raw = df_raw.merge(base_munic_2020_meio_amb.drop(['UF', 'Cod UF','Mun', 'Faixa_pop', 'Regiao'], axis=1), how='inner', left_on='CD_MUN', right_on='CodMun')
df_raw = df_raw.merge(base_munic_2020_gest_riscos.drop(['UF', 'Cod UF','Mun', 'Faixa_pop', 'Regiao'], axis=1), how='inner', left_on='CD_MUN', right_on='CodMun')
df_raw['ano'] = 2020
df_raw.head()
```

```python
df_exp_ama_legal = df_exportacao[['Year', 'Month', 'Country', 'City', 'US$ FOB', 'Net Weight', 'SH2 Description']].merge(df_uf_mun[['City', 'NM_MUN', 'CD_MUN']], how='inner', on='City')
df_exp_ama_legal.drop_duplicates(inplace=True)
df_exp_ama_legal.head()
```

```python
df_exp_best_item = df_exp_ama_legal.loc[df_exp_ama_legal['Year']==2020].groupby(['Year', 'SH2 Description', 'CD_MUN']).agg(sum_best_item_dolar=('US$ FOB', 'sum')).reset_index().sort_values(['CD_MUN', 'sum_best_item_dolar', 'Year'], ascending=False)

munics = df_exp_best_item.CD_MUN.unique()
df_exp_best_item_unique = pd.DataFrame(columns=['Year','SH2 Description', 'CD_MUN',	'sum_best_item_dolar'])

for munic in munics:
   df_aux = df_exp_best_item.loc[(df_exp_best_item['CD_MUN']==munic)].nlargest(1, 'sum_best_item_dolar')
   df_exp_best_item_unique = pd.concat([df_aux, df_exp_best_item_unique])

df_exp_best_item_unique.head()
```

```python
df_exp_total = df_exp_ama_legal.loc[df_exp_ama_legal['Year']==2020].groupby(['Year', 'CD_MUN']).agg(sum_total_dolar=('US$ FOB', 'sum')).reset_index()
df_exp_total.head()
```

```python
df_raw = df_raw.merge(df_exp_total, how='inner', left_on=['CD_MUN', 'ano'], right_on=['CD_MUN', 'Year']).drop(['Year'], axis=1)
df_raw = df_raw.merge(df_exp_best_item_unique, how='inner', left_on=['CD_MUN', 'ano'], right_on=['CD_MUN', 'Year']).drop(['Year'], axis=1)
df_raw.drop(['CodMun_x', 'CodMun_y', 'CodMun'], axis=1, inplace=True)
df_raw.head()
```
